```python
import re
import PyPDF2 as pdf
import pandas as pd
```

```python
# Import pdf
store = pdf.PdfFileReader('Source/top-baby-suburbs.pdf')
text_data_0 = store.getPage(0).extractText()
text_data_1 = store.getPage(1).extractText()
```

```python
# Regex Pattern Creation
pattern = re.compile("(\d{3}[\D][\w ]+)\n")
col_headers = re.compile("201\d").findall(text_data_0)
col_headers.pop(0)

page0 = pattern.findall(text_data_0)
page0.pop(0)
print(len(page0))
                        
page1 = pattern.findall(text_data_1)
page1.pop(0)
print(len(page1))

page0.extend(page1)

x =[]
for sl in page0:
    x.append([sl[3:],sl[:3]])
```

```python
line =[]
data_list = []
for idx, l in enumerate(x):
    if (idx) % 8 == 0 and idx != 0:
        data_list.append(line)
        line = []
    line.append(l)
data_list.append(line)
```

```python
# Create dataframe
data_table = pd.DataFrame(data_list, columns=col_headers)
data_table.index = data_table.index +1

data_table
```

```python
# For Extracting
for year in range(2011, 2019):
    test = []
    for row in data_table[str(year)]:
        test.append([row[0],row[1]])    
    
    # Extract data 
    table = pd.DataFrame(test, columns=['Suburb ' + str(year),'Number-' + str(year)])
    table.index = table.index + 1
    table.to_csv("Output Files/" + str(year) + "-file.csv")

```
